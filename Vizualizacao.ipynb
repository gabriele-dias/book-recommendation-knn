```python
import pandas as pd
tab_cruzada = pd.read_csv('tab_pronta.csv')
tab_cruzada.describe()
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o arquivo que você salvou
df = pd.read_csv('tab_pronta.csv')

# Renomear para nomes fáceis de digitar (Sugestão)
df.rename(columns={
    'Book-Title': 'titulo',
    'Book-Author': 'autor',
    'Book-Rating': 'nota',
    'User-ID': 'id_usuario',
    'Book-Rating' : 'avaliacao'
    
}, inplace=True)

# Ver as primeiras linhas
df.head()
```

```python
# Use o nome que o pandas carregou do CSV
tab_cruzada = tab_cruzada[tab_cruzada['Book-Rating'] > 0]

# Agora sim, verifique o resultado
print(f"Total de linhas agora: {tab_cruzada.shape[0]}")
```

```python
#analise grafica 

plt.title('analisando a avaliacao')
sns.boxplot(data=tab_cruzada, x='Book-Rating');
```

```python


# Realizando a análise agrupada
Analise = tab_cruzada.groupby(by=['Book-Title']).agg(
    Quantidade=('Book-Title', 'count'),
    Media=('Book-Rating', 'mean'),
    Maxima=('Book-Rating', 'max'),
    Minima=('Book-Rating', 'min'),
    Mediana=('Book-Rating', 'median')
)

# Verificando o resultado
Analise.head()
```

```python
#livros que tiveram mais avaliações

Analise.sort_values('Quantidade', ascending=False).head()
```

```python
Analise.sort_values(['Media', 'Quantidade'], ascending=False).head()
```

```python
import plotly.express as px
import plotly.io as pio

# Isso vai abrir o gráfico no seu navegador padrão
pio.renderers.default = "browser"

fig = px.scatter(data_frame=Analise, 
                 x='Quantidade', 
                 y='Media', 
                 title='Media Vs Qntd de titulos')
fig.show()
```

```python
Analise['Quantidade'].describe()
```

```python
def Classificacao_Quantidade(Quantidade):
    if int(Quantidade) <= 5:
        return '1-5 Avaliações'
    elif int(Quantidade) <= 10:
        return '6-10 Avaliações'
    elif int(Quantidade) <= 100:
        return '11-100 Avaliações'
    else:
        return '> 101 Avaliações'

# 1. Aplica a sua função e conta os valores
Pizza = Analise['Quantidade'].apply(Classificacao_Quantidade).value_counts()

# 2. Transforma em DataFrame e reseta o index
# O reset_index() vai criar colunas chamadas 'Quantidade' e 'count' (ou 'index')
Pizza = pd.DataFrame(Pizza).reset_index()

# 3. Verifique os nomes das colunas para não errar no gráfico
# print(Pizza.columns) # Descomente para conferir se der erro

# 4. Plot (Ajustado para os nomes que o Pandas gera)
px.pie(
    data_frame=Pizza,
    names='Quantidade',  # A categoria (texto)
    values='count',       # O valor numérico (contagem)
    title='Proporção de Livros por Volume de Avaliações'
).show()
```

```python
# Cria a contagem e já transforma em DataFrame com colunas limpas
Analise_Ano = tab_cruzada['Year-Of-Publication'].value_counts().sort_index().reset_index()

# Renomeia as colunas para evitar erros de KeyError
Analise_Ano.columns = ['Ano', 'Total_Livros']

# Verifica como ficou
Analise_Ano.head()
```

```python
import matplotlib.pyplot as plt

# Filtra entre 1990 e 2023
filtro_anos = Analise_Ano[(Analise_Ano['Ano'] > 1990) & (Analise_Ano['Ano'] < 2023)]

plt.figure(figsize=(12, 6))
plt.bar(filtro_anos['Ano'], filtro_anos['Total_Livros'], color='royalblue')

plt.title('Publicação de livros por ano (1990 - 2023)')
plt.xlabel('Ano de Publicação')
plt.ylabel('Quantidade de Livros')
plt.xticks(rotation=45)
plt.show()
```

```python
# Autores com mais avaliações na base de dados 

Top_Autores = tab_cruzada.groupby(by='Book-Author').agg(
    Quantidade=('Book-Rating', 'count'),
    Media=('Book-Rating', 'mean')
).sort_values('Quantidade', ascending=False).head(10)

# Verificando o resultado
Top_Autores
```

```python
# Regiao
analise_regiao = tab_cruzada['Pais'].value_counts(normalize=True).cumsum() * 100
print(analise_regiao.head(10))
```

```python
# O normalize=True calcula a proporção (ex: 0.71)
# O .cumsum() vai somando essas proporções acumuladas
resultado_acumulado = tab_cruzada['Pais'].value_counts(normalize=True).cumsum()

print(resultado_acumulado.head(10))
```
