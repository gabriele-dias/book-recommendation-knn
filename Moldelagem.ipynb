```python
##MODELAGEM DE DADOS 

#1 cruzamento

tab_cruzada = Livros.merge(Avaliacao, how= 'inner', on='ISBN')


#2 cruzamento 

tab_cruzada = tab_cruzada.merge(Usuarios, how='inner', on='User-ID')

#dIMENSAO
tab_cruzada.shape
```

```python
#Tratamento de texto
def Extrair_Pais(Regiao):
    # Tudo que está dentro da função deve ter exatamente o mesmo recuo (4 espaços ou 1 Tab)
    Registro = Regiao.split(',')
    fracao = Registro[-1]
    return fracao.strip()

tab_cruzada['Pais'] = tab_cruzada['Location'].apply (Extrair_Pais)

#verificando

tab_cruzada.head()
```

```python
#renomer as colunas 

# Renomear respeitando as maiúsculas originais que aparecem no seu print
tab_cruzada.rename(columns={
    'Book-Title': 'titulo_do_livro',
    'Book-Author': 'autor_do_livro',
    'Year-Of-Publication': 'ano_de_publicacao',
    'Publisher': 'editora',
    'User-ID': 'id_do_usuario',
    'Book-Rating': 'Avaliacao',
    'Location': 'Localizacao',
    'Age': 'idade'
}, inplace=True)

# Agora sim, verifique se mudou
print(tab_cruzada.columns)

#verificar
tab_cruzada.columns
```
